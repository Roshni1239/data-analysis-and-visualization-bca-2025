```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    '2020': [20, 34, 30, 35, 27],
    '2025': [25, 30, 45, 32, 28]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 6))
```

```python
for i in range(len(df)):
    plt.plot([0, 1], [df.loc[i, '2020'], df.loc[i, '2025']], marker='o', label=df.loc[i, 'Product'])
    plt.text(-0.05, df.loc[i, '2020'], f"{df.loc[i, 'Product']} {df.loc[i, '2020']}", va='center', ha='right')
    plt.text(1.05, df.loc[i, '2025'], f"{df.loc[i, 'Product']} {df.loc[i, '2025']}", va='center', ha='left')
```

```python
plt.xticks([0, 1], ['2020', '2025'])
plt.title("Slope Chart: Product Sales from 2020 to 2025")
plt.ylabel("Sales")
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [20, 35, 30, 35, 27]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))
plt.hlines(y=df['Category'], xmin=0, xmax=df['Value'], color='skyblue', linewidth=2)
plt.plot(df['Value'], df['Category'], 'o', color='blue')
plt.title('Lollipop Chart Example')
plt.xlabel('Value')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 5))
plt.vlines(x=df['Category'], ymin=0, ymax=df['Value'], color='lightgreen', linewidth=2)
plt.plot(df['Category'], df['Value'], 'o', color='green')

plt.title('Vertical Lollipop Chart')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(42)
data = {
    'Employee': [f'Emp{i}' for i in range(1, 11)],
    'Performance_Score': np.random.normal(70, 10, 10)
}

df = pd.DataFrame(data)
```

```python
average_score = df['Performance_Score'].mean()
df['Score_Diff'] = df['Performance_Score'] - average_score
```

```python
df['Color'] = ['green' if x > 0 else 'red' for x in df['Score_Diff']]
```

```python
df_sorted = df.sort_values('Score_Diff')

plt.figure(figsize=(10, 6))
plt.barh(df_sorted['Employee'], df_sorted['Score_Diff'], color=df_sorted['Color'])

plt.axvline(0, color='black', linewidth=0.8)
plt.title('Diverging Bar Chart: Performance vs. Average')
plt.xlabel('Difference from Average Score')
plt.show()
```
